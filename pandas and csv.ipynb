```python
s = ""
b = open('data.csv','w')
b.write("Name,Age,Address\n")
n = int(input("enter n: "))
for i in range(n):
    name = input("name: ")
    age = int(input("age: "))
    add = input("add: ")
    info = f"{name},{age},{add}\n"
    s =s+info
    
b.write(s)
b.close()
```
